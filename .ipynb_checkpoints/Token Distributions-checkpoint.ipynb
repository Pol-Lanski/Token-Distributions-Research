## Liquidity Mining

### Terminology

Liquidity mining is a form of **Generalized Mining**. 

As per Jack Bruckman from CoinFund: 
> Generalized mining (also:“keeping”, “supply-side services”, and “mining 2.0”) refers to the cryptoeconomic games implemented by decentralized protocols that users can play to earn cryptocurrency-denominated compensation.

> Decentralized networks typically require hardware, software, or manual services in order to function. However, they cannot provide such services themselves (for reasons of both economics and decentralization) and therefore rely on users, investors, and other third parties for service provision. Decentralized networks make use of cryptoeconomic mechanisms to incentivize third parties, creating compensatory opportunities. Examples of such services include transaction processing, staking, computational resource provision, software or Merkle Mining, content curation, market making, and more.

It can be considered "Proof-of-work" outside of the base layer. The "work" itself can be anything that the network needs and wants to pay for and can be proved/verified programatically - as opposed to a Bounty Program. The "Proof-of-Work" consensus mechanism of Bitcoin and Ethereum is what pays for securing the network, and is built-in on the base layer.

**Liquidity mining** then, is a sub-set of generalized mining that incentivizes liquidity provision.

Finally **Yield Farming** is an investment strategy that seeks to gain revenue (yield) through generalized mining, of which liquidity mining is the most common. 


### Projects applying liquidity mining

From Idex in 2017, Livepeer in 2018 to Synthetix in 2019, liquidity mining is not a new idea. But when Compound launched its liquidity mining program in June 2020, it all exploded.

### The different "liquidity minings"

Liquidity mining is "proof-of-work" in which the work rewarded is liquidity provision, but apart from incentivating such work there might be other reasons of second order on why a project would run a liquidity mining program. The difference will depend on the whether the token distributed is a governance or utility token, to which stakeholders the tokens need to be distributed, whether the goal is really to decentralize governance or to gain users of a protocol, maybe reward previous supporters, provide exit liquidity to investors or development teams or just want to increase the price of their existing token. Most likely, it is a combination of the above, and there are more legitimate and sustainable distributions than others. 

[Dmitriy Berenzon](https://medium.com/bollinger-investment-group/liquidity-mining-a-user-centric-token-distribution-strategy-1d05c5174641) from Bollinger Investment Group differenciates between three categories:

1. Fair Launches
2. Programmatic Decentralization
3. Growth Marketing

#### Fair Launches
"Fair launches" aim to distribute **the majority** of the tokens *without* a sale.
Note that the concept of "fairness" is contested.
and some named Fair launches benefitted particular types of investors that were supposed to put at an equal level with the community supporter. This is ***especially*** true for those Fair Launches tied to liquidity mining and governance tokens. With a lot of capital, these investors can farm a lot of the rewards and end up controlling the majority of the governance. 
Defining characteristics: Majority of tokens distributed

#### Programatic decentralization
This programs aim to gradually distribute power to the community, over a long period of time and often with incentives to stay for the long run.
Defining characteristics: long period of distribution and relatively high token distribution.

#### Growth Marketing
They incentivize a particular behavior over a defined period of time. They often do not change the governance structure and are targeted at attracting new users or more holders to their token.
Short period of time and small number of tokens used for the incentive program.
